# ***CLASSWORK: Import numpy and show different methods in it***

```python
# importing numpy library
import numpy as np
```

```python
# creating single dimensional array
np.array([10,50,60,70])
```

```python
# creating array with zeroes
np.zeros((10,5))
```

```python
# creating two single dimensional array
a1=np.array([1,2,3,4])
a2=np.array([5,6,7,8])
np.sum(a1)
```

```python
# summation of two array
a1+a2
```

```python
# creating a multi dimensional array
a3=np.array([[1,2,3,4],[6,7,8,9]])
```

```python
# sum of array columnwise
a3.sum(axis=0)
```

```python
# sum of array rowwise
a3.sum(axis=1)
```

```python
# arange method to print number in a certain range
np.arange(50)
```

```python
# linspace method
np.linspace(0,100,10)
```

```python
# checking shape if array
a3.shape
```

```python
a1=np.array([1,2,3,4])
a2=np.array([5,6,7,8])
```

```python
# creating vertical stack
np.vstack((a1,a2))
```

```python
# creating horizontal stack
np.hstack((a1,a2))
```

```python
# creating column stack
np.column_stack((a1,a2))
```

```python
# finding mean of array 
np.mean(a1)
```

```python
# finding standard deviation of array a1
np.std(a1)
```

```python
# finding median of array a1
np.median(a1)
```

```python
# saving and loading an array
np.save("first_array.npy",a1)
```

```python
# loading the saved array
np.load("first_array.npy")
```

```python
# converting array into a list
print("the type of a1 is",type(a1))
a3=list(a1)
print("after converting the type of a1 is",type(a3))
```
