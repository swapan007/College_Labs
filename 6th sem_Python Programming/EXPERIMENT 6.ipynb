##### **Experiment 6 :**Write a program to compute the x and y cordinates for points on a sin curve and plot the points using matplotlib.

###### Method 1

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.linspace(0,5*np.pi)
y=np.sin(x)
```

```python
plt.plot(x,y,color="g",linewidth=1)
plt.title("Sin(x)")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.show()
```

##### Method 2

```python
x = np.arange(0, 3 * np.pi, 0.2)
y = np.sin(x)
print("Plotting the points using matplotlib:")
plt.plot(x, y)
plt.title("Sin(x)")
plt.show()
```
