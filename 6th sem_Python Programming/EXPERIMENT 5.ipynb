# **EXPERIMENT 5:** Write a NumPy program to create a 10x10 matrix, in which the elements on the borders will be equal to 1, and inside 0.
## **NAME:** SWAPAN CHETRI 

```python
# importing the required library
import numpy as np
```

```python
# taking the input from the user
input_1=int(input("Enter the no of rows:: "))
input_2=int(input("Enter the no of columns:: "))
row=input_1-1
# creating a zeros matrix of 10x10
array=np.zeros((input_1,input_2))
```

```python
# logic
for i in range(input_1):
  array[0][i]=1
  array[i][0]=1
  array[row][i]=1
  array[i][row]=1
```

```python
# printing array
print(array)
```

```python
# creating a ones matrix of 10x10
array_2=np.ones((input_1,input_2))
```
