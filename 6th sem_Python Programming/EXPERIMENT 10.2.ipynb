Write a Python program to draw a line using given axis values taken from a text file, with suitable label in the x
axis, y axis and a title.nTest Data: test.txt 1 2n2 4n3 1

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = range(1, 30)
Y = [value * 2 for value in X]
print("Values of X: ")
print(*range(1,30)) 
print("Values of Y (twice of X): ")
print(Y)
```

```python
np.savetxt("array.txt",(X,Y))
```

```python
data=np.loadtxt("array.txt")
```

```python
data
```

```python
plt.plot(data[0],data[1])
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.grid()
plt.title('Line with label in the x axis, y axis')
plt.show()
```
